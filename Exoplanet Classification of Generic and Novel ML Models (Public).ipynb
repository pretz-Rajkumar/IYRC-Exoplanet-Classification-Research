```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
X_train = np.load("/content/drive/MyDrive/IYRC Research/IYRC Submission Material/x_train_res.npy")
y_train = np.load("/content/drive/MyDrive/IYRC Research/IYRC Submission Material/y_train_res.npy")
X_test = np.load("/content/drive/MyDrive/IYRC Research/IYRC Submission Material/x_test.npy")
y_test = np.load("/content/drive/MyDrive/IYRC Research/IYRC Submission Material/y_test.npy")
```

```python
# Train a Perception Model
from sklearn.linear_model import Perceptron  #Perception model
ppn = Perceptron(eta0=0.1, random_state=1)  #Constructor: to create object of machine learning model
ppn.fit(X_train, y_train)  #ppn.fit(): learning to predict

y_pred = ppn.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report, balanced_accuracy_score
print("Accuracy: %.3f" % accuracy_score(y_test, y_pred))
print("Balanced Accuracy: %.3f" % balanced_accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))


from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='Perceptron Algorithm')
display.plot()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=100.0, solver='lbfgs', multi_class='ovr')
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report, balanced_accuracy_score
print("Balanced Accuracy: %.3f" % balanced_accuracy_score(y_test, y_pred))
print("Accuracy: %.3f" % accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report
print("Balanced Accuracy: %.3f" % balanced_accuracy_score(y_test, y_pred))
print("Accuracy: %.3f" % accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(100,50))  #Constructor: to create object of machine learning model, eta0=learning rate
mlp.fit(X_train, y_train)  #ppn.fit(): learning to predict

# Make Predictions via the Predict Method
y_pred = mlp.predict(X_test)

#  Calculate classification accuracy of perceptron
print("Accuracy: %.3f" % accuracy_score(y_test, y_pred))
print("Balanced Accuracy: %.3f" % balanced_accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(criterion='entropy', max_depth=8, random_state=1)
tree_model.fit(X_train, y_train)

y_pred = tree_model.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report
print(classification_report(y_test, y_pred))
print("Balanced Accuracy: %.3f" % balanced_accuracy_score(y_test, y_pred))


from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display2 = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='Decision Tree Classifier')
display.plot()
display2.plot()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
random_model = RandomForestClassifier(n_estimators=220)
random_model.fit(X_train, y_train)

y_pred = random_model.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report
print(classification_report(y_test, y_pred))
print("Balanced Accuracy: %.3f" % balanced_accuracy_score(y_test, y_pred))
```
